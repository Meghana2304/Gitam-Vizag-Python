### Data Cleaning
- Whenever you have a lot of missing records values in the given dataset then you can isnull()
- isnull() function will notify any missing values in the dataframe

```python
# Entire dataset
df.isnull().sum()
```

```python
# Only column missing values
df['Season'].isnull().sum()
```

- ***dropna()*** This Function allows you drop all or some rows that have missing values
- ***fillna()*** This function allows you to replace the rows that have missing values with the values what you want to pass

```python
len(df)
```

```python
# Dropping rows with 1 null value
new_df = df.dropna(axis = 0, how = 'any')
```

### Visualizing data
- An interesting way to display Dataframe
- with matplotlib

```python
import matplotlib.pyplot as plt
a1 = [1,2,3,4]
a2 = [5,6,7,8]
plt.figure(figsize=(5,5))
plt.plot(a1,a2,color='green')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
divisions = ['A','B','C','D']
marks = [79,82,75,63]
avg_marks = [65,72,78,89]
width = 0.70
plt.bar(divisions,marks,color = 'blue', width = 0.3)
plt.show()
```

```python
plt.barh(divisions,marks,color = 'blue')
plt.show()
```

```python
lang = ['Python','Java','C','C++']
rating = [75,70,55,65]
plt.pie(rating,labels = lang,shadow = True, startangle = 90, autopct = '%1.1f%%')
plt.legend(loc = 'best')
plt.show()
```
